```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image_path = r"E:\PRO\WBC_Segment\segmentation_WBC\Dataset 1\011.bmp"  # Replace with the actual path to your image
image = cv2.imread(image_path)

# Convert BGR to RGB (OpenCV uses BGR by default, but Matplotlib uses RGB)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```

```python
# Convert BGR to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')  # Use the 'gray' colormap for grayscale images
plt.axis('off')
plt.show()
```

```python
# Apply histogram equalization
enhanced_image = cv2.equalizeHist(gray_image)

# Display the enhanced image
plt.imshow(enhanced_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Apply Canny edge detection
edges = cv2.Canny(blurred_image, 500, 150)

# Display the blurred image and edges
plt.figure(figsize=(10, 5))

# plt.subplot(1, 2, 1)
#plt.title("Blurred Image")
plt.imshow(blurred_image, cmap='gray')
plt.axis('off')
plt.show()
```
